```python
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras import layers,models
import matplotlib.pyplot as plt
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory("PlantVillage")
```

```python
dataset.take(1)
for i,l in dataset.take(1):
    print(i.shape)
    for j in range(16):
        plt.subplot(4,4,j+1)
        plt.imshow(i[j].numpy().astype('uint8'))
```

```python
train_df_size = len(dataset)*0.08
train_df_size
```

```python
train_df = dataset.take(52)
test_df = dataset.skip(52)
dataset.class_names
```

```python
model = models.Sequential([
    layers.Conv2D(100,(3,3),activation='relu',input_shape=(256,256,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(100,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(100,activation='relu'),
    layers.Dense(16)
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
h = model.fit(train_df,epochs=2)
```

```python
h.history
```
